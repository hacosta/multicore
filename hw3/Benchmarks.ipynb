```python
import matplotlib.pyplot as plt
import subprocess
import time
%matplotlib inline
```

```python
subprocess.check_call(['make'])
```

```python
threads = [1, 2, 4, 8, 16]
times = []
for T in threads:
    start = time.time()
    subprocess.check_call(['./matrix_mult', 'tests/testcase0_1', 'tests/testcase0_2', str(T)])
    elapsed = time.time() - start
    times.append(elapsed)

assert(len(threads) == len(times))
plt.bar(threads, times)
```

```python
threads = range(1, 16)
times = []
for T in threads:
    start = time.time()
    subprocess.check_call(['./matrix_mult', 'tests1000/testcase0_1', 'tests1000/testcase0_2', str(T)])
    elapsed = time.time() - start
    times.append(elapsed)

assert(len(threads) == len(times))
plt.bar(threads, times)
```
